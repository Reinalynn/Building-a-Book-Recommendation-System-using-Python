```python
import pandas as pd
```

```python
books_10k = pd.read_csv('Data/books10k.csv')
books_10k.head()
```

```python
reviews_10k = pd.read_csv('Data/reviews10k_grouped.csv')
reviews_10k.head()
```

```python
books_10k.dtypes
```

```python
# discovered that reviews use goodreads book id as book id
# renaming column in reviews_df3
reviews_10k.rename(columns = {'book_id': 'goodreads_book_id'}, inplace = True)
reviews_10k.head()
```

```python
# add review_text from reviews2, combining by book_id
reviews_full = pd.merge(books_10k, reviews_10k, on = 'goodreads_book_id', how = 'left')
reviews_full.head()
```

```python
reviews_full.columns
```

```python
reviews_full = reviews_full.drop(['authors_x', 'original_title_x', 'image_url', 'small_image_url'], axis = 1)
```

```python
print(reviews_full.shape)
```

```python
reviews_full.columns
```

```python
reviews_full.to_csv('Data/reviews10k_grouped_full.csv', index = False)
```
