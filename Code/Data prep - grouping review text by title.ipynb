```python
ds2 = ds[['book_id', 'authors', 'original_title', 'review_text']]
ds2.head()
```

```python
ds2.dtypes
```

```python
# https://stackoverflow.com/questions/39303912/tfidfvectorizer-in-scikit-learn-valueerror-np-nan-is-an-invalid-document
data_types_dict = {'authors': str, 'original_title': str, 'review_text': str}
ds3 = ds2.astype(data_types_dict)
ds3.dtypes
```

```python
# group by goodreads_book_id, authors, and original_title and concat review_text (1 row/book)
# https://stackoverflow.com/questions/27298178/concatenate-strings-from-several-rows-using-pandas-groupby
ds3['review_text'] = ds3[['book_id','authors','original_title', 'review_text']].groupby(['book_id','authors', 'original_title'])['review_text'].transform(lambda x: ','.join(x))
ds3.head()
```

```python
print(ds3.shape)
```

```python
ds4 = ds3[['book_id','authors','original_title','review_text']].drop_duplicates()
```

```python
print(ds4.shape)
```
