```python
import pandas as pd
import gzip
```

```python
# use python generator to iterate through books2 and reviews2
def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        yield eval(l)
```

```python
books2 = (parse('goodreads_books.json.gz'))
type(books2)
```

```python
next(books2)
```

```python
reviews2 = (parse('goodreads_reviews_dedup.json.gz'))
rev_lst = list(reviews2)
len(rev_lst)
```

```python
rev_cols = ('user_id', 'book_id', 'review_id', 'rating', 'review_text', 'date_added', 'date_updated', 'read_at', 'started_at', 
           'n_votes', 'n_comments')
reviews_df = pd.DataFrame(rev_lst, columns = rev_cols)
```

```python
reviews_df.head()
```

```python
reviews_df.dtypes
```

```python
type(reviews_df)
```

```python
# https://www.tutorialspoint.com/change-data-type-for-one-or-more-columns-in-pandas-dataframe-1
data_types_dict = {'book_id': int}
reviews_df3 = reviews_df.astype(data_types_dict)
reviews_df3.dtypes
```

```python
books_10k = pd.read_csv('books10k.csv')
books_10k.head()
```

```python
books_10k.dtypes
```

```python
# discovered that reviews use goodreads book id as book id
# renaming column in reviews_df3
reviews_df3.rename(columns = {'book_id': 'goodreads_book_id'}, inplace = True)
reviews_df3.head()
```

```python
reviews_df3.loc[reviews_df3['goodreads_book_id'] == 2767052]
```

```python
# add review_text from reviews2, combining by book_id
reviews_10k = pd.merge(books_10k, reviews_df3, on = 'goodreads_book_id', how = 'left')
reviews_10k.head()
```

```python
reviews_10k.to_csv('reviews_10k.csv', index = False)
```
